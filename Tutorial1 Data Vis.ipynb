## Multiple bar chart
* this part is based on [this](https://github.com/DataForScience/DataViz/blob/master/1.%20Matplotlib.ipynb)


```python
dataLA = [20, 14, 23]
dataSF = [11, 18, 28]

pos = np.arange(0, 5, 2)  # set position of bar on X axis
width = 0.5  # set width of bar
```

## Pie chart
* pie is not good for visualization, compared to bar chart

```python
dataA = [17, 18, 20, 22, 24]
labelsA = ["A1", "A2", "A3", "A4", "A5"]
dataB = [20, 20, 19, 21, 20]
labelsB = ["B1", "B2", "B3", "B4", "B5"]
dataC = [24, 22, 20, 18, 17]
labelsC = ["C1", "C2", "C3", "C4", "C5"]

# we can plot charts in one graph
fig, axs = plt.subplots(nrows=2, ncols=3)

axs[0][0].pie(dataA, labels=labelsA)
axs[0][1].pie(dataB, labels=labelsB)
axs[0][2].pie(dataC, labels=labelsC)

# we have specify the color of the bar
colors = plt.rcParams["axes.prop_cycle"].by_key()["color"]
axs[1][0].bar(labelsA, dataA, color=colors)
axs[1][1].bar(labelsB, dataB, color=colors)
axs[1][2].bar(labelsC, dataC, color=colors)
```

## Seaborn
* this part is based on [this](https://github.com/DataForScience/DataViz/blob/master/4.%20Seaborn.ipynb)

```python
# load libraries
import seaborn as sns
```

```python
# load data
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

#### data property
- `total_bill`: Total bill amount.
- `tip`: Tip amount.
- `sex`: Gender of the customer.
- `smoker`: Whether the customer is a smoker.
- `day`: Day of the week.
- `time`: Whether it’s lunch or dinner.
- `size`: Number of people in the party.

```python
# we can specify the style of the graph, decomment the following line
# plt.style.use("fivethirtyeight")
# scatter plot using seaborn
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# you can find the style of the graph by using the following command
# try another style
plt.style.available
```

### Bubble plot
* seaborn can easily make complex chart like bubble plot 

```python
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    kind="scatter",
    hue="sex",
    size="size",
    # style="smoker",  # we can also add different marker, decomment the this line
)
```

```python
# we can also easily draw multiple plots in one graph
sns.relplot(
    x="total_bill",
    y="tip",
    data=tips,
    kind="scatter",
    hue="sex",
    size="size",
    # col="smoker",row="day",style="time",  # decomment this line
    # row_order=["Thur", "Fri", "Sat", "Sun"],  # decomment this line
)
```

### Catergorical plot


```python
sns.catplot(
    x="sex",  # we can change this to "smoker" to see the difference
    y="total_bill",  # we can change this to "tip" to see the difference
    data=tips,
    kind="strip",
    hue="day",
    jitter=False,
    dodge=True,
)
```

```python
sns.catplot(
    x="sex",  # we can change this to "smoker" to see the difference
    y="total_bill",  # we can change this to "tip" to see the difference
    data=tips,
    kind="box",  # we can change this to "violin" to see the difference
    hue="day",
)
```
