```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, (x - 5) ** 2, label=f"L = $(w - 5)^2$")
plt.plot(x, np.abs(x), label=f"$L_1$ = |w|")

plt.plot(x, (x - 5) ** 2 + np.abs(x), label=f"L = $(w - 5)^2$ + |w|")
plt.xlim(-10, 10)
plt.ylim(0, 10)
plt.legend()
```

```python
plt.plot(x, (x - 5) ** 2, label=f"L = $(w - 5)^2$")
plt.plot(x, x**2, label=f"$L_2 = w^2$")

plt.plot(x, (x - 5) ** 2 + np.abs(x), label=f"L = $(w - 5)^2 + w^2$")
plt.xlim(-10, 10)
plt.ylim(0, 10)
plt.legend()
```
